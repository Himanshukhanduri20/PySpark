```python
data={'Time':[],
      'Date':[],
      'Data_source_host_name':[],
      'log_category':[],
      'attack_type':[],
      'description':[],
      'seviority':[],
      'src_ip':[],
      'user_name':[],
      'destination_ip':[],
      'desitination_host':[],
      'rule_message':[],
      'Data_source_ip':[],
      'destination_port':[]
      }
```

```python
import re
zz=[]
for i in new_data:
  r=re.findall("(>[a-zA-Z]+\s*\d+\s*[\d+:]+)",i)
  time="".join(re.findall("([\d+:]+\d)",str(r)))
  data['Time'].append(time)


  r="".join(r)
  date=r.replace(time,"")
  date=date.replace('>',"")
  data['Date'].append(date)

  Data_source_host_name="".join(re.findall("([a-zA-Z]+[-][a-zA-Z]+[-][a-zA-Z]+[-]\d+)",i))
  data['Data_source_host_name'].append(Data_source_host_name)

  r="".join(re.findall("(cat=[a-zA-Z]+)",i))
  log_category=r.replace('cat=',"")
  data['log_category'].append(str(log_category))

  r="".join(re.findall("(#011name=[a-zA-Z:\s]+)",i))
  attack_type=r.replace("#011name=","")
  data['attack_type'].append(attack_type)


  r="".join(re.findall("(011desc=[a-zA-Z\s\/.]+)",i))
  description=r.replace("011desc=","")
  data['description'].append(description)

  r="".join(re.findall("(011sev=[\d]+)",i))
  seviority =r.replace("011sev=","")
  data['seviority'].append(seviority)

  r=re.findall("#011src=[0-9]{1,3}[.][0-9]{1,3}[.][0-9]{1,3}[.][0-9]{1,3}",str(i))
  src_ip = "".join(re.findall("[0-9]{1,3}[.][0-9]{1,3}[.][0-9]{1,3}[.][0-9]{1,3}",str(r)))
  data['src_ip'].append(src_ip)

  r="".join(re.findall("(011usrName=[a-zA-Z \s]+)",i))
  user_name=r.replace("011usrName=","")
  data['user_name'].append(user_name)

  r=re.findall("(#011target=[0-9 . ( ) a-zA-Z ]+)",str(i))
  destination_ip = "".join(re.findall("[0-9]{1,3}[.][0-9]{1,3}[.][0-9]{1,3}[.][0-9]{1,3}",str(r)))
  data['destination_ip'].append(destination_ip)

  r="".join(r)
  desitination_host=r.replace("#011target="+destination_ip,"")
  data['desitination_host'].append(desitination_host)

  r="".join(re.findall("(#011msg=[0-9 . \/ a-zA-Z ]+)",str(i)))
  rule_message =r.replace("#011msg=","")
  data['rule_message'].append(rule_message)

  r=re.findall("#011TrendMicroDsScannerIp=[0-9]{1,3}[.][0-9]{1,3}[.][0-9]{1,3}[.][0-9]{1,3}",str(i))
  Data_source_ip="".join(re.findall("[0-9]{1,3}[.][0-9]{1,3}[.][0-9]{1,3}[.][0-9]{1,3}",str(r)))
  data['Data_source_ip'].append(Data_source_ip)

  r=re.findall("(TrendMicroDsTargetPortList=[\d*;]*\d*)",str(i))
  destination_port=";".join(re.findall('(\d+)',str(r)))
  data['destination_port'].append(destination_port) 







 

```
